```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
import datetime
import calendar

# Importing the dataset
df = pd.read_csv('test.txt', sep="\t", error_bad_lines=False)
s = df['item_type'].str.split(',').apply(Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'item_type'
del df['item_type']
df = df.join(s)
df['item_type'] = df['item_type'].apply(lambda x: x.split('_')[-1].replace('\'','').replace(']',''))
df = df[df['item_type']!='null'].reset_index()
del df['index']
del df['role_id']
del df['rk']


df['t_when'] = df['t_when'].apply(lambda x: calendar.timegm(datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S").utctimetuple())) 

d = defaultdict(LabelEncoder)
pd.Categorical(df['app_channel'])
pd.Categorical(df['item_type'])

le = LabelEncoder()
fit = df.apply(lambda x: d[x.name].fit_transform(x) if (x.name == ('item_type')) else x)
df = df.apply(lambda x: d[x.name].fit_transform(x) if (x.name in ('item_type', 'app_channel')) else x )
df
```
